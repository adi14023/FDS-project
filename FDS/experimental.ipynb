```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
le=LabelEncoder()
```

```python
le_an = LabelEncoder()
le_s1 = LabelEncoder()
le_an.fit(df['AnimalName'])
le_s1.fit(df['symptoms1'])
df['symptoms1'] = le_s1.transform(df['symptoms1'])
df['AnimalName'] = le_an.transform(df['AnimalName'])
```

```python
le_y=LabelEncoder()
df['Dangerous'] = le_y.fit_transform(df['Dangerous'])
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state=90)
```

```python
model_svm = SVC(kernel = 'linear', random_state = 0)
model_svm.fit(X_train, y_train)
```

```python
y_pred = model_svm.predict(X_test)
print(y_pred)
```

```python
accuracy_svm = accuracy_score(y_test, y_pred)
print(accuracy_svm)
```

```python
classification_rep_svm = classification_report(y_test, y_pred)
print(classification_rep_svm)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print(classification_report(y_test, y_pred))
```
